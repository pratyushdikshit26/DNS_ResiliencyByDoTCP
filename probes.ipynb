```python
def display_map_for_probes():
    fig = px.scatter_mapbox(probes_df,
                        lat=probes_df.latitude,
                        lon=probes_df.longitude,
                        color = probes_df[" "], 
                        zoom=1)
    
    
    fig.update_layout(legend=dict(
    orientation="h",
    yanchor="bottom",
    y=1.02,
    xanchor="right",
    x=1
    ))
    
    fig.show()        

    fig.write_image("plots/probes.pdf", width=980, height=550)

```

```python
df = pd.read_csv("tmp_csv_files/prb_to_resolver_counts.csv")
```

```python
df
```

```python
def create_ip_version_column(row):
    addr = ipaddress.ip_address(row['dst_addr'])
    if addr.version == 4:
        return "IPv4"
    else:
        return "IPv6"
    
def create_ip_info_column(row):
    addr = ipaddress.ip_address(row['dst_addr'])
    if addr.is_private:
        return "Private"
    else: 
        return "Public"
                
```

```python
df["ip_info"] = df.apply(create_ip_info_column, axis=1)
df["ip_version"] = df.apply(create_ip_version_column, axis=1)
```

```python
pd.crosstab(df["ip_info"], df["ip_version"], margins = True)
```

```python
pd.crosstab(df["ip_info"], df["ip_version"], normalize="all", margins=True)
```
