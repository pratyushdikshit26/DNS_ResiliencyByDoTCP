```python
from ripe.atlas.cousteau import AtlasResultsRequest
from ripe.atlas.sagan import DnsResult

from Log import Log
import Variables as Variables
from ripe_wrapper.database.hop_result_table import HopResultTable
from ripe_wrapper.database.measurement_table import MeasurementTable
from ripe_wrapper.database.measurement_probe_table import MeasurementProbeTable
from ripe_wrapper.database.result_table import ResultTable
from ripe_wrapper.database.database_table import DatabaseTable
import time
from listing import MyMeasurementRequest
from datetime import datetime
```

```python
def handle_traceroute_measurement(results):
    measurement_probe_table = MeasurementProbeTable()
    hop_result_table = HopResultTable()
    entries = results
    for entry_i in range(len(entries)):
        original_result = entries[entry_i]
        measurement_probe_table.save_measurement_probe(original_result)
        hop_result_table.save_hop_result(original_result)

def handle_dns_measurement(results):
    measurement_probe_table = MeasurementProbeTable()
    result_table = ResultTable()
    entries = results
    for entry_i in range(len(entries)):
        original_result = entries[entry_i]
        measurement_probe_table.save_measurement_probe(original_result)
        parsed_result = DnsResult.get(original_result)
        result_table.save_result(original_result, parsed_result)
```
