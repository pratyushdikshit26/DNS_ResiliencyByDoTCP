```python
import csv 
import time
tsv_file = open("ip2asn-v4-u32.tsv")
read_tsv = csv.reader(tsv_file, delimiter="\t")
for row in read_tsv:
    item = {
        'number': row[2],
        'name': row[4]
    }
    filters = {'number': row[2]}
    if not as_table.exists(filters):
        try:
            as_table.add_item(item)
        except: 
            print(item)
```
